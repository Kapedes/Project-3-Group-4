```python
# Import dependencies
import pandas as pd
import sqlite3 as sql
import json
```

```python
# Connect to the SQLite database
conn = sql.connect('movie.sqlite')

# Create a cursor
cursor = conn.cursor()

# Execute a SQL query to retrieve table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Fetch all table names
table_names = cursor.fetchall()

# Print the table names
print("Tables in the database:")
for table in table_names:
    print(table[0])
```

```python
# Print the table names and their columns
print("Tables and columns in the database:")

# Iterate through each table name fetched
for table in table_names:
     
    # Extract the name of the current table
    table_name = table[0]
    print(f"\nTable: {table_name}")
    
    # Execute PRAGMA command to retrieve column information
    cursor.execute(f"PRAGMA table_info({table_name});")
    
    # Fetch all column information
    columns = cursor.fetchall()
    print(columns)
```

```python
# Execute a SQL query to retrieve table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Initialize an empty dictionary to store the fetched data
data = {}

# Fetch all table names from the database
table_names = cursor.fetchall()

# Iterate through each table name fetched
for table in table_names:
    # Extract the name of the current table
    current_table = table[0]
    
    # Execute SQL query to select all rows from the current table
    cursor.execute(f"SELECT * FROM {current_table}")
    
    # Fetch all rows returned by the query
    rows = cursor.fetchall()
    
    # Initialize an empty list to store dictionaries representing rows of data
    data_list = []
    
    # Iterate through each row fetched from the current table
    for row in rows:
        # Create a dictionary representing the current row of data
        # by zipping column names with row values
        data_list.append(dict(zip([column[0] for column in cursor.description], row)))
    
    # Store the list of dictionaries representing rows of data
    # for the current table in the data dictionary using the table name as the key
    data[current_table] = data_list

# Display data
data
```

```python
# Convert data to JSON format
json_data = json.dumps(data, indent=4)

# Write JSON data to a file
with open('static/js/output.json', 'w') as json_file:
    json_file.write(json_data)
```
